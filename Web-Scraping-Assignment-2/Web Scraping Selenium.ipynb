#### 1. Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
# Setting up the WebDriver
service = Service(r"chromedriver.exe")
driver = webdriver.Chrome(service=service)
```

```python
url = 'https://www.shine.com/'
driver.get(url) # Opening the webpage

# Clicking the search icon to load the search options
search = driver.find_element(By.XPATH, '//span[@class="iconH-zoom mr-30"]')
search.click()
time.sleep(3) # Waiting for the page to load

# Entering Job Designation as Data Analyst
designation_input = "Data Analyst"
designation = driver.find_element(By.XPATH, '//input[@id="id_q"]')
designation.send_keys(designation_input)

# Entering Job Location as Bangalore
location_input = "Bangalore"
location = driver.find_element(By.XPATH, '//input[@id="id_loc"]')
location.send_keys(location_input)
time.sleep(2)

blr = driver.find_element(By.XPATH, '//li[@index="id_loc_0"]')
blr.click()
time.sleep(3)

# Clicking on the Search button
search_btn = driver.find_element(By.XPATH, '//button[@type="submit"]')
search_btn.click()

# Initializing empty lists to store the data
job_title = []
job_location = []
company_name = []
experience_reqd = []

# Scraping first 10 job titles
title = driver.find_elements(By.XPATH, '//h2[@itemprop="name"]')
for i in title[:10]:
    job_title.append(i.text)

# Scraping first 10 job locations    
location = driver.find_elements(By.XPATH, '//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in location[:10]:
    job_location.append(i.text)
    
# Scraping first 10 company names   
company = driver.find_elements(By.XPATH, '//div[@class="jobCard_jobCard_cName__mYnow"]')
for i in company[:10]:
    company_name.append(i.text)

# Scraping first 10 job experience required 
experience = driver.find_elements(By.XPATH, '//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in experience[:10]:
    experience_reqd.append(i.text)
    
# Creating dataframe   
df = pd.DataFrame({'Job Title': job_title,
                   'Job Location': job_location,
                   'Company Name': company_name,
                   'Experience Required': experience_reqd
                  })
df
```

```python
driver.quit()
```

#### 4. Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
1. Brand
2. ProductDescription
3. Price

```python
# Setting up the WebDriver
service = Service(r"chromedriver.exe")
driver = webdriver.Chrome(service=service)
```

```python
url = 'https://www.flipkart.com/'
driver.get(url) # Opening the webpage

# Clicking the X button to close the popup
x_button = driver.find_element(By.XPATH, '//button[@class="_2KpZ6l _2doB4z"]')
x_button.click()

# Finding the search bar and entering the search query for sunglasses
product_input = 'sunglasses'
product = driver.find_element(By.CLASS_NAME, "_3704LK")
product.send_keys(product_input)

# Clicking the search button
search = driver.find_element(By.CLASS_NAME, "L0Z3Pu")
search.click()
time.sleep(3) # Giving time to load the page


#Scraping first 100 sunglass data

sunglasses_data = []
desired_product_count = 100
items_per_page = 40

# Calculating the number of pages to scrape
pages_to_scrape = (desired_product_count + items_per_page - 1) // items_per_page

while len(sunglasses_data) < desired_product_count and pages_to_scrape > 0:
    # Locate the parent container for each item
    item_containers = driver.find_elements(By.XPATH, '//div[@class="_1xHGtK _373qXS"]')

    # Extracting the data for each item
    for container in item_containers:
        brand = container.find_element(By.XPATH, './/div[@class="_2WkVRV"]')
        try:
            desc = container.find_element(By.XPATH, './/a[@class="IRpwTa"]')
            description = desc.text
        except:
            description = None
        price = container.find_element(By.XPATH, './/div[@class="_30jeq3"]')

        data = {
            'brand': brand.text,
            'description': description,
            'price': price.text
        }
        sunglasses_data.append(data)

        # Checking if the desired item count is reached
        if len(sunglasses_data) == desired_product_count:
            break

    # Moving to the next page
    next_page_link = driver.find_element(By.XPATH, '//a[contains(., "Next")]')
    next_page_link.click()
    time.sleep(3) # Waiting for next page to load

    pages_to_scrape -= 1
    
# Creating the dataframe    
df = pd.DataFrame(sunglasses_data)
df
```

```python
driver.quit()
```

#### 5. Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link:
https://www.flipkart.com/apple-iphone-11-black-64-gb/productreviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART


```python
# Setting up the WebDriver
service = Service(r"chromedriver.exe")
driver = webdriver.Chrome(service=service)
```

```python
url = 'https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&market'
driver.get(url) # Opening the webpage

# Scraping the first 100 reviews
ratings = []
review_title = []
review_description = []

desired_review_count = 100
reviews_per_page = 10

# Calculating the number of pages to scrape
pages_to_scrape = (desired_review_count + reviews_per_page - 1) // reviews_per_page

for page in range(pages_to_scrape):
    
    # Extracting the ratings
    rating = driver.find_elements(By.XPATH, '//div[@class="_3LWZlK _1BLPMq" or @class="_3LWZlK _1rdVr6 _1BLPMq"]')
    for i in rating:
        ratings.append(i.text)
        
    # Extracting the review summary  
    titles = driver.find_elements(By.XPATH, '//p[@class="_2-N8zT"]')
    for i in titles:
        review_title.append(i.text)
    
    # Clicking on "READ MORE" to expand the full review
    try:
        read_more = driver.find_element(By.XPATH, '//span[@class="_1BWGvX"]')
        driver.execute_script("arguments[0].click();", read_more)
        time.sleep(3) # Waiting for the review to load
    except:
        time.sleep(1)
    # Extracting the full review
    desc = driver.find_elements(By.XPATH, '//div[@class="t-ZTKy"]')
    for i in desc:
        review_description.append(i.text)
        
    # Moving to the next page
    next_page_link = driver.find_element(By.XPATH, '//a[contains(., "Next")]')
    next_page_link.click()
    time.sleep(7) # Waiting for next page to load
    
    
df = pd.DataFrame({'Rating': ratings, 'Review Summary': review_title, 'Full Review': review_description})
df
```

```python
driver.quit()
```

#### 7. Go to webpage https://www.amazon.in/ Enter “Laptop” in the search field and then click the search icon. Then set CPU Type filter to “Intel Core i7” as shown in the below image: After setting the filters scrape first 10 laptops data. You have to scrape 3 attributes for each laptop:
1. Title
2. Ratings
3. Price

```python
# Setting up the WebDriver
service = Service(r"chromedriver.exe")
driver = webdriver.Chrome(service=service)
```

```python
url = 'https://www.amazon.in/'
driver.get(url)

product_input = 'Laptop'
product = driver.find_element(By.XPATH, '//input[@class="nav-input nav-progressive-attribute"]')
product.send_keys(product_input)

# Clicking the search button
search = driver.find_element(By.XPATH, '//input[@type="submit"]')
search.click()
time.sleep(3) # Giving time to load the page

# Clicking checkbox for filtering
filter_checkbox = driver.find_element(By.XPATH, '/html/body/div[1]/div[2]/div[1]/div[2]/div/div[3]/span/div[1]/div/div/div[6]/ul[6]/span[9]/li/span/a/div')
filter_checkbox.click()

laptop_titles = []
laptop_ratings = []
laptop_price = []

# Scraping the first 10 titles
titles = driver.find_elements(By.XPATH, '//span[@class="a-size-medium a-color-base a-text-normal"]')
for i in titles[:10]:
    laptop_titles.append(i.text)

# Scraping the first 10 ratings
ratings = driver.find_elements(By.XPATH, '//span[contains(@class, "a-icon-alt")]')
for i in ratings[:10]:
    full_rating = i.get_attribute('innerHTML') # Getting the full rating data
    r = full_rating.split()[0] # Extracting the rating
    laptop_ratings.append(r)

# Scraping the first 10 laptop prices
price = driver.find_elements(By.CLASS_NAME, "a-price-whole")
for i in price[:10]:
    laptop_price.append(i.text)
    

# Creating the dataframe
df = pd.DataFrame({'Title': laptop_titles, 'Rating (out of 5)': laptop_ratings, 'Price (in ₹)': laptop_price})
df
```

#### 8. Write a python program to scrape data for Top 1000 Quotes of All Time. The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on TopQuotes
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
# Setting up the WebDriver
service = Service(r"chromedriver.exe")
driver = webdriver.Chrome(service=service)
```

```python
url = 'https://www.azquotes.com//'
driver.get(url)

# Clicking the navigation button to open the list of navigation options
nav_button = driver.find_element(By.XPATH, '//a[@id="showhide"]')
nav_button.click()
time.sleep(3) # Waiting to load the navigation options

# Navigating to the top quote page by clicking the Top Quotes option
top_quote_page = driver.find_element(By.XPATH, '//a[contains(text(), "Top Quotes")]')
top_quote_page.click()
time.sleep(3) # Waiting to load the page

# Initializing empty lists to store dara
quote_texts = []
authors = []
quote_types = []

#Scraping data from all 10 pages
for page in range(10):
    #Extracting entire div for each quote
    quotes = driver.find_elements(By.XPATH, '//div[@class="wrap-block"]')
    
    for quote in quotes:
        # Scraping data for the 100 quotes on each page
        quote_text = quote.find_element(By.XPATH, './/a[@class="title"]').text
        author = quote.find_element(By.XPATH, './/div[@class="author"]').text
        quote_type = quote.find_element(By.XPATH, './/div[@class="tags"]').text
        
        quote_texts.append(quote_text)
        authors.append(author)
        quote_types.append(quote_type)
        
    if page == 9: # Breaking the loop on the last page
        break
        
    # Going to the next page
    next_page = driver.find_element(By.XPATH, '//a[@rel="nofollow" and contains(text(), "Next")]')
    next_page.click()
    time.sleep(7) #Waiting for page to load
        
# Create a dataframe from the scraped data
data = {
    'Quote': quote_texts,
    'Author': authors,
    'Type of Quote': quote_types
}
df = pd.DataFrame(data)
df
```

```python
driver.quit()
```
